```python
import pandas as pd
import matplotlib.pyplot as plt
import os
#print(os.getcwd())
#print(os.listdir())
df = pd.read_csv("Data/DigitSpan/emotivDigitSpan_AFFECT.csv") 

df['timestamp'] = pd.to_numeric(df['timestamp'])
df['timestamp'] = (df['timestamp'] - df['timestamp'].iloc[0])
#print(df)
```

```python
metrics = ["engagement", "excitement", "stress", "relaxation", "interest"]
colors = ["blue", "red", "orange", "green", "purple"]

plt.figure(figsize=(12, 6))

for metric, color in zip(metrics, colors):
    plt.plot(df['timestamp'], df[metric], label=metric.capitalize(), color=color, linewidth=2)

plt.xlabel("Time (seconds)")
plt.ylabel("Metric Value (0-1)")
plt.title("Digit Span Test Emotional Metrics Over Time")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
df2 = pd.read_csv("Data/DigitSpan/emotivDigitSpan_PAD.csv") 

df2['timestamp'] = pd.to_numeric(df2['timestamp'])
df2['timestamp'] = (df2['timestamp'] - df2['timestamp'].iloc[0])

metrics = ["P", "A", "D"]
colors = ["blue", "red", "green"]

plt.figure(figsize=(12, 6))

for metric, color in zip(metrics, colors):
    plt.plot(df2['timestamp'], df2[metric], label=metric.capitalize(), color=color, linewidth=2)

plt.xlabel("Time (seconds)")
plt.ylabel("Metric Value (-1 to 1)")
plt.title("Digit Span Test Emotional Metrics Over Time")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

df2 = pd.read_csv("DigitSpan/log.csv")
df2['Timestamp'] = pd.to_numeric(df2['Timestamp'])
df2 = df2.dropna(subset=['Timestamp'])
df2 = df2.sort_values(by="Timestamp")

action_mapping = {'Fail': 0, 'Success': 1}
df2['Outcome'] = df2['Outcome'].map(action_mapping)
df_part1 = df2.iloc[0:9].copy()
df_part2 = df2.iloc[9:19].copy()
df_part1['Timestamp'] = df_part1['Timestamp'] - df_part1['Timestamp'].iloc[0]
df_part2['Timestamp'] = df_part2['Timestamp'] - df_part2['Timestamp'].iloc[0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), subplot_kw={'projection': '3d'})
ax1.set_box_aspect(None, zoom=0.85)
ax2.set_box_aspect(None, zoom=0.85)
ax1.scatter(df_part1['Timestamp'], df_part1['Level'], df_part1['Outcome'], c=df_part1['Outcome'], cmap="bwr", marker="o", edgecolors="black", s=80, alpha=1.0)
ax1.set_xlabel("Relative Time (seconds)")
ax1.set_ylabel("Level")
ax1.set_zlabel("Outcome (0=Fail, 1=Success)")
ax1.set_title("3D Level Progression (Indexes 1-9)")
ax2.scatter(df_part2['Timestamp'], df_part2['Level'], df_part2['Outcome'], c=df_part2['Outcome'], cmap="bwr", marker="o", edgecolors="black", s=80, alpha=1.0)
ax2.set_xlabel("Relative Time (seconds)")
ax2.set_ylabel("Level")
ax2.set_zlabel("Outcome (0=Fail, 1=Success)")
ax2.set_title("3D Level Progression (Indexes 10-19)")

plt.tight_layout()
plt.show()
```

```python

df2 = pd.read_csv("CorsiBlock/log.csv")

df2 = df2[pd.to_numeric(df2['Timestamp'], errors='coerce').notna()]
df2 = df2[pd.to_numeric(df2['Level'], errors='coerce').notna()]
df2['Timestamp'] = pd.to_numeric(df2['Timestamp'])
df2['Level'] = pd.to_numeric(df2['Level'])
df2.reset_index(drop=True, inplace=True)

df2['Outcome'] = df2['Outcome'].map({'Fail': 0, 'Success': 1})
df_part1 = df2.iloc[:3].copy()
df_part2 = df2.iloc[3:].copy() 
df_part1['Timestamp'] -= df_part1['Timestamp'].iloc[0]
df_part2['Timestamp'] -= df_part2['Timestamp'].iloc[0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), subplot_kw={'projection': '3d'})
ax1.set_box_aspect(None, zoom=0.85)
ax2.set_box_aspect(None, zoom=0.85)
ax1.scatter(df_part1['Timestamp'], df_part1['Level'], df_part1['Outcome'], c=df_part1['Outcome'], cmap="bwr", marker="o", edgecolors="black", s=80, alpha=1.0)
ax1.set_xlabel("Relative Time (seconds)")
ax1.set_ylabel("Level")
ax1.set_zlabel("Outcome (0=Fail, 1=Success)")
ax1.set_title("3D Level Progression (Indexes 1-3)")
ax2.scatter(df_part2['Timestamp'], df_part2['Level'], df_part2['Outcome'], c=df_part2['Outcome'], cmap="bwr", marker="o", edgecolors="black", s=80, alpha=1.0)
ax2.set_xlabel("Relative Time (seconds)")
ax2.set_ylabel("Level")
ax2.set_zlabel("Outcome (0=Fail, 1=Success)")
ax2.set_title("3D Level Progression (Indexes 4+)")

plt.tight_layout()
plt.show()
```
